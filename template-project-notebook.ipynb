```python
name = df['grade'].tolist()
tropical_cyclones =[]
for i in name:
    if 'Tropical Cyclone of TS intensity or higher'in i:
        tropical_cyclones.append(i)
print(len(tropical_cyclones))#количество тропических циклонов в наборе данных
```

```python
pressure = df['Central pressure'].tolist()
max_value = pressure [0]
min_value = pressure [0]
for i in pressure:
    if i > max_value:
        max_value = i
    if i < min_value:
        min_value = i
print('Максимальное давление:', max_value)
print('Минимальное давление:', min_value)
#максимальное и минимальное давление за период с 1951 по 2022 год
```

```python
data_list = df.values.tolist()
tropical_cyclones = []
for row in data_list:
    for value in row:
        if 'Tropical Cyclone of TS intensity or higher' in str(value):
            tropical_cyclones.append(row)
            break
for cyclone in tropical_cyclones:
    print(cyclone)
```
